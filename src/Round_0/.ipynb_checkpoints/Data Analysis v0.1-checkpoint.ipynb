```python


# Check for stationarity using ADF test
result = adfuller(rainforest_resin_data['mid_price'])
print("ADF Statistic:", result[0])
print("p-value:", result[1])
#p-value is zero, imlpying

# Fit AR(1) model
model_ar1 = AutoReg(rainforest_resin_data['mid_price'], lags=1)
model_ar1_fitted = model_ar1.fit()
print("AR(1) Coefficient:", model_ar1_fitted.params[1])

# Fit ARMA model
model_arma = ARIMA(rainforest_resin_data['mid_price'], order=(1, 0, 1))  # AR(1) + MA(1)
model_arma_fitted = model_arma.fit()
print(model_arma_fitted.summary())
```

```python
import numpy as np
import pandas as pd
from pykalman import KalmanFilter
import matplotlib.pyplot as plt

# Example data (replace with your actual data)
data = rainforest_resin_data['mid_price']  # Replace with your actual data

# Initialize the Kalman Filter
# Set initial state mean (using the mean of the data as an initial guess for the long-term mean)
initial_state_mean = data.mean()

# Define the Kalman Filter for the Ornstein-Uhlenbeck process
kf = KalmanFilter(initial_state_mean=initial_state_mean, n_dim_obs=1, 
                  transition_matrices=[1],  # Identity matrix (no change in state)
                  observation_matrices=[1],  # Observation matrix (direct observation of the state)
                  transition_covariance=1e-1,  # Process noise (adjust for mean reversion)
                  observation_covariance=1e-1)  # Observation noise

# Fit the model using Expectation-Maximization (EM) algorithm
kf = kf.em(data, n_iter=5)  # 5 iterations of EM algorithm to estimate parameters

# Apply the Kalman filter to estimate the underlying state
(filtered_state_means, filtered_state_covariances) = kf.filter(data)

# The filtered state means represent the smoothed estimates of the process
# These correspond to the estimates of the underlying process, including mean reversion

# Extract the Kalman filter parameters
# For an OU process, theta (mean reversion rate) and sigma (volatility) are part of the filter parameters
print("Estimated mean (mu):", kf.initial_state_mean)
print("Estimated process noise (sigma):", kf.transition_covariance)
print("Estimated observation noise (observation covariance):", kf.observation_covariance)

# Plot the original data and the filtered (OU) process
plt.figure(figsize=(10, 6))
plt.plot(data, label='Original Data', color='blue')
plt.plot(filtered_state_means, label='Filtered (OU) Process', linestyle='--', color='red')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Mid Price')
plt.title('Ornstein-Uhlenbeck Process Filtering')
plt.grid(True)
plt.show()
```
