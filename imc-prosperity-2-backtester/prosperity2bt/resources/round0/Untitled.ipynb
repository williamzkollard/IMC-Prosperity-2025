```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("prices_round_0_day_-1.csv", delimiter = ";")
```

```python
df["profit_and_loss"] = 0.0 
df["day"] = -2
df = df.fillna("")
```

```python
df
```

```python
df_single = df.astype(str).agg(";".join, axis=1).to_frame(name="merged")
```

```python

data_rows = df.astype(str).agg(";".join, axis=1)

# Convert header to a string
header_row = ";".join(df.columns)

# Combine header and data into one list
all_rows = [header_row] + data_rows.tolist()

# Save to CSV (4000x1 style with semicolon delimiter)
with open("output.csv", "w") as f:
    for row in all_rows:
        f.write(row + "\n")
```
